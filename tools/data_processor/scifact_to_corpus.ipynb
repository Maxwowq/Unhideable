```python
import json
```

```python
ORIGNAL_DATA_PATH = "../../data/scifact/original_corpus.jsonl"
OUTPUT_PATH= "../../data/scifact/corpus.jsonl"
```

```python
docs = []
with open(ORIGNAL_DATA_PATH, "r") as f:
    for line in f:
        data = json.loads(line.strip())
        docs.append({
            "_id":data["doc_id"],
            "title":data["title"],
            "text":"".join(data["abstract"])
        })

with open(OUTPUT_PATH, "w") as f:
    for doc in docs:
        json.dump(doc, f, ensure_ascii=False)
        f.write("\n")
```
