```python
import json
```

```python
with open("/mnt/data1/workspace/zms/LeakDojo/data/chatdoctor/HealthCareMagic-100k.json", "r") as f:
    data = json.load(f)
```

```python
data[0]
```

```python
rag_text_outputs = []
```

```python
for i in range(len(data)):
    rag_text_outputs.append(
    {
        "_id": str(i),
        "title": "",
        "text": "input: " + data[i]["input"] + "\noutput: " + data[i]["output"],
        "metadata":{}
    }
    )
```

```python
rag_text_outputs[0]
```

```python
with open("corpus.jsonl", "w", encoding="utf-8") as out:
    for item in rag_text_outputs:
        out.write(json.dumps(item, ensure_ascii=False) + "\n")
```
